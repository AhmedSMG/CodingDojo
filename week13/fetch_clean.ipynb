```python
import numpy as np
import pandas as pd
import yfinance as yf

from tqdm import tqdm

import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# This csv file taken from nasdaq contains all available tickers
security_all = pd.read_csv("./nasdaq_screener.csv",
                           usecols=["Symbol", "Name", "Market Cap", "Country", "Volume", "Sector", "Industry"],
                           )
security_all.head()
```

```python
security_all["Sector"].fillna("Other", inplace=True)
security_all["Industry"].fillna("Other", inplace=True)
security_all
```

```python
start_date = pd.Timestamp("2005-01-01")
end_date = pd.Timestamp("2021-12-31")
print(f"#of security {security_all.shape[0]}")
```

```python
# Fetch historical data
df = yf.download(security_all["Symbol"].to_list(), start_date, end_date)["Adj Close"]
```

```python
# save all hostrical data as csv file
df.to_csv("./all_historical_data.csv")
```

```python
df
```

```python
# Cleaning Histroical Data
cols = []
min_days = df.isna().sum().min()
for col in df.columns:
    days = df[col].isna().sum()
    percent = days / df.shape[0]
    print(f"{col:<6}: {days:<7}, {percent:.3f} %")
    if days > min_days:
        cols.append(col)
```

```python
min_days
```

```python
cols
```

```python
df_clean = df.copy()
```

```python
# Drop unwanted tickers
df_clean = df_clean.drop(columns=cols)
df_clean
```

```python
# Drop records with NaN
df_clean = df_clean.dropna(axis=0)
df_clean
```

```python
df_clean.to_csv("./clean_historical_data.csv")
```

```python
tickers = df_clean.columns
tickers
```

```python
# Select only the security in our clean historical data
security = security_all.query("Symbol in @df_clean.columns")
security
```

```python
security.isna().sum()
```

```python
# filling the missing data using yfinance lib
for col, k in zip(["Country", "Sector", "Industry"], ["country", "sector", "industry"]):
    idx = security.loc[security[col].isna()].index
    print(f"{col}: {k}")
    for i in tqdm(idx):
        security.at[i, col] = yf.Ticker(security.loc[i, "Symbol"]).get_info().get(k, None)
```
