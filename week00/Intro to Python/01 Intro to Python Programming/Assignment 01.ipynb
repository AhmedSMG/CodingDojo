2. Multiples of Five - Print all the multiples of 5 from 5 to 1,000

3. Counting, the Dojo Way - Print integers 1 to 100. If divisible by 5, print "Coding" instead. If divisible by 10, print "Coding Dojo".

4. Whoa. That Sucker's Huge - Add odd integers from 0 to 500,000 and print the final sum.

5. Countdown by Fours - Print positive numbers starting at 2018, counting down by fours.

6. Flexible Counter (optional) - Set three variables: lowNum, highNum, mult. Starting at lowNum and going through highNum, print only the integers that are a multiple of mult. For example, if lowNum=2, highNum=9, and mult=3, the loop should print 3, 6, 9 (on successive lines)

```python
# Q1 -> Print all integers from 0 to 150. Hint: use a for loop and range
print(list(range(0, 150 + 1)))
```

```python
# Q2 -> Print all the multiples of 5 from 5 to 1,000
print(list(range(5, 1000 + 1, 5)))
```

```python
# Q3 -> 
# Print integers 1 to 100. 
# If divisible by 5, print "Coding" instead
# If divisible by 10, print "Coding Dojo".
list(map(lambda x: "Coding" * (x % 5 == 0) + " Dojo" * (x % 10 == 0) or str(x), range(1, 100 + 1)))
```

```python
# Q4 -> Add odd integers from 0 to 500,000 and print the final sum.
print(sum(range(1, 500_000 + 1, 2)))
```

```python
# Q5 -> Print positive numbers starting at 2018, counting down by fours
print(list(range(2018, 0, -4)))
```

```python
# Q6 -> lowNum, highNum, mult.
# Starting at lowNum and going through highNum,
# print only the integers that are a multiple of mult
lowNum, highNum, mult = 50, 100, 4
print(list(range(lowNum if lowNum > mult else mult, highNum + 1, mult)))
```
