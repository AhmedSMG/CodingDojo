```python
from datetime import datetime
import numpy as np
import pandas as pd
import pandas_datareader as pdr
```

```python
tickers = ["AAPL", "MSFT", "TWTR", "IBM", "^GSPC"]
start = datetime(2015, 12, 1)
end = datetime(2021, 1, 1)

stocks = pdr.get_data_yahoo(tickers, start, end, interval="m")
stocks.head()
```

```python
log_returns = np.log(stocks["Adj Close"]/stocks["Adj Close"].shift())
log_returns
```

```python
cov = log_returns.cov()
var = log_returns["^GSPC"].var()
```

```python
beta = cov.loc["AAPL", "^GSPC"] / var
```

```python
beta
```

```python
risk_free_return = 0.0138
market_return = 0.105
expected_return = risk_free_return + beta * (market_return - risk_free_return)
```

```python
expected_return
```

```python
beta * market_return
```
