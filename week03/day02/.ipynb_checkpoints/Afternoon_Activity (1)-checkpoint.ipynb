# Goal:

#### Calculate the line of best fit for some data

#### Calculate Perasons correlation

#### Measure effect sizes

```python
import pandas 
import pandas as pd

import numpy 
import numpy as np

import matplotlib.pyplot as plt
```

#### Generating Data to use in our learning example

```python
n = 35 

np.random.seed(0)

# independent variable random sample
# Normalized income per household
x = np.random.uniform(size = n) * 15
e = np.random.uniform(size = n) * 0.5
# Calculate y as "number of children" = f("house hold income")

b1 = a = 0.5 

b2 = m = 0.22

y = b1 + b2*x + e
```

#### Calculate the line of best fit

using the three following equations





```python
tally = pandas.DataFrame(data={'x':x,'y':y})
tally.index.name = 'Element #'
tally_df = tally
tally_df.head(6)

tally_df = tally_df.sort_values('y')
tally_df.head()
```

```python

# we use bar to refer to averages
x_bar = tally_df.x.mean()
y_bar = tally_df.y.mean()

print(f" x_bar is {x_bar}")
print(f" y_bar is {y_bar}")

print(tally_df.head(4))

# lets add the terms for our equation
tally_df
tally_df['x_div'] = tally_df['x'] - x_bar
tally_df['y_div'] = tally_df['y'] - y_bar
tally_df['x_div_sqr'] = tally_df['x_div']**2

print(tally_df.head(4))
```

```python

# We need to sovle for m_hat and b_hat

# we use hats to refer to estimates
# m is the slope of our line of best fit
m_hat = (tally_df['x_div'] * tally_df['y_div']).sum() / tally_df['x_div_sqr'].sum()
m_hat

# b is the intercept, b_hat is the intercept estimated from our data
b_hat = y_bar - (m_hat*x_bar)
b_hat

print("\n")
print(f"m_hat is {m_hat}")
print(f"b_hat is {b_hat}")

# these are our y_estimates per our model. 
tally_df['y_hat'] = b_hat + (tally_df['x']*m_hat)
tally_df.head(4)
```

#### Demonstrating the effect of noise ('epsilon')


```python

#calculate epsilon
tally_df['e'] = tally_df['y_hat'] - tally_df['y']
```

#### Generating Figures for raw data and estimates

```python


fig = tally_df.plot(x='x',y='y_hat',kind='scatter')
fig.legend(['fitted estimates from Econometric Equation'])
plt.title('Econometric Equation')
fig.plot(tally_df['x'].values[[0,-1]],
         tally_df['y_hat'].values[[0,-1]])


plt.xlabel('Household income (Normalized)')
plt.ylabel('Children per Household')


fig_2 = tally_df.plot(x='x',y='y',kind='scatter')
fig_2.plot(tally_df['x'].values[[0,-1]],
         tally_df['y_hat'].values[[0,-1]])

plt.title('Raw Data')

plt.xlabel('Household income (Normalized)')
plt.ylabel('Children per Household')

fig_2.legend(['line of best fit'])
```

```python
plt.scatter(x=tally_df['x'],
            y=tally_df['y'].values)

plt.scatter(
    x = tally_df['x'],
    y = (tally_df['x'].values * m_hat) + b_hat
)

plt.plot(tally_df[['x']],
         (tally_df['x'].values * m_hat) + b_hat)
# is also
plt.plot(tally_df[['x']],
         tally_df[['y_hat']])
```
